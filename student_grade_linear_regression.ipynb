# Student Grade Prediction using Linear Regression

### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

### Loading the datasets

```python
df = pd.read_csv('data/student_info.csv')
df
```

### Showing first 5 datas

```python
df.head()
```

### Getting the information of the data

```python
df.info()
```

```python
df.describe()
```

### Finding the missing values in the data

```python
df.isnull().sum()
```

### Normalizing the null data

```python
df2 = df.fillna(df.mean())
df2
```

### Plotting a graph

```python
plt.scatter(x=df2.study_hours, y=df2.student_marks, color='red', marker='*')
plt.show()
```

```python
sns.heatmap(df2.corr(), annot=True, linewidths=3)
plt.show()
```

```python
sns.barplot(data=df2, x=df2['study_hours'], y=df2['student_marks'])
plt.show()
```

### Spliting the dataset

```python
X = df2.drop("student_marks", axis='columns')
y = df2.drop("study_hours", axis='columns')
print(f'Shape of x: {X.shape}')
print(f'Shape of Y: {y.shape}')
```

```python
sns.pairplot(df2, kind='reg')
```

### Spliting the data for training and testing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=51)
print("-"*20+"Training Data Shape"+"-"*20)
print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of y_train: {y_train.shape}")
print("-"*20+"Testing Data Shape"+"-"*20)
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

## Selecting a model and train it
### Linear Regression
#### y = m*x+c

```python
lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

### Creating a DataFrame for easy analysis

```python
pd.DataFrame(np.c_[X_test, y_test, y_pred], columns=["study_hours", "student_original_marks", "student_predicted_marks"])
```

### Checking the Accuracy of our Model

```python
lr.score(X_test, y_test)
```

```python
plt.scatter(X_train, y_train)
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_train, lr.predict(X_train), color='red')
```

## Saving Our Machine Learning Model

```python
import joblib
joblib.dump(lr, "Student_grade_prediction_model.pkl")
```
